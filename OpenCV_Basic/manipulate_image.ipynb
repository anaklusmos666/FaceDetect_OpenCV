```python
#libraries
import cv2
import numpy as np
import pandas as pd
```

# Read Image

```python
img_color = cv2.imread('catto.jpg', 1) #reads color image
img_gray = cv2.imread('catto.jpg', 0) #reads grayscale
img_original = cv2.imread('catto.jpg', -1) #reads original image
```

# Show Image using Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(img_gray, cmap='gray', interpolation='bicubic')
plt.xticks([]) #hide label & ticks
plt.yticks([]) #hide label & ticks
plt.show()
```

```python
b, g, r = cv2.split(img_color) #cv reads image in bgr order. matplotlib reads in rgb order
temp_img = cv2.merge([r, g, b]) #converting image BGR > RGB

#distorted image, bgr
plt.subplot(121)
plt.imshow(img_color)
plt.xticks([]); plt.yticks([])

#original image, rgb
plt.subplot(122)
plt.imshow(temp_img)
plt.xticks([]); plt.yticks([])
```

```python
cv2.imshow('catto', img_color)
cv2.waitKey(1000)
cv2.destroyAllWindows()
```

# add line

```python
img_line = cv2.line(img_color, (0,0), (255, 255), (0, 0, 0), 2)
cv2.imshow('edit', img_line)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
